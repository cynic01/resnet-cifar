```python
import matplotlib.pyplot as plt
```

```python
num_workers = list(range(0, 20+1, 4))
time = [16.913351496999894, 0.7979606640019483, 0.9282079270064969, 1.0108642480067829, 1.1118243409996467, 1.2158078610027587]
```

```python
plt.plot(num_workers, time)
plt.plot(num_workers, time, 'o')
plt.xticks(num_workers)
for i, v in zip(num_workers, time):
    plt.gca().text(i, v+0.4, "%.3f" %v, ha="center")
plt.ylim(0, 18)
plt.show()
```
